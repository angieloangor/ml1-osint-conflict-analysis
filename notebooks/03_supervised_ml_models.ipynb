## Limitaciones de la fase supervisada

- Las etiquetas generadas son `weak labels` heurísticas basadas en palabras clave y temas existentes.
- Puede existir un fuerte desbalance de clases, en particular `other` y `escalation` dominan el corpus.
- `google_news_rss` es la fuente dominante del corpus, lo que puede sesgar los modelos hacia su estilo editorial y contenido.
- No se usa `NASA FIRMS` ni `OpenSky` como texto; esos datos sólo deben considerarse como contexto operacional en dashboards separados.
- Esta aproximación es válida como prueba académica para ML1, pero no sustituye una taxonomía validada ni ground truth militar.
