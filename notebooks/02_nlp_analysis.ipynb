## Metodología del análisis NLP

1. Se carga el corpus textualmente validado en `data/dataset_nlp.csv`.
2. Se realiza limpieza, tokenización, eliminación de stopwords y lematización con NLTK.
3. Se calcula TF-IDF para extraer representaciones semánticas de cada noticia.
4. Se aplican NMF para inferir tópicos latentes y KMeans sobre embeddings semánticos para agrupar noticias similares.
5. Se generan métricas de calidad de cluster (silhouette) y se guarda un reporte estructurado.
6. Se identifican automáticamente temas relacionados con Israel-Irán con palabras clave y se analizan agrupamientos temporales.
